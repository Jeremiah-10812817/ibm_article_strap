```python
#import packages
from bs4 import BeautifulSoup
import requests
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
```

```python
url="https://www.ibm.com/cloud/learn/natural-language-processing"

# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).text

# Parse the html content
soup = BeautifulSoup(html_content, "lxml")
 # print the parsed data of html
soup.prettify()
```

```python
# get article content
#remove html tags
#remove /n from text
data=soup.find("div", class_="ibm-lh-body")
data = soup.getText()
data= " ".join(data.split())
data
```

```python
# Creating text parser using tokenization
parser = PlaintextParser.from_string(data, Tokenizer("english"))

from sumy.summarizers.text_rank import TextRankSummarizer

# Summarize using sumy TextRank
summarizer = TextRankSummarizer()
summary = summarizer(parser.document, 2)

text_summary = ""
for sentence in summary:
  text_summary += str(sentence)

print(text_summary)                
    
```
